```python
import os

def download_file(color, year):
    for i in range(1, 13):
        month = f'{i:02d}'
        os.makedirs(f'data/raw/{color}/{year}/{month}', exist_ok=True)
        url = f'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/{color}/{color}_tripdata_{year}-{month}.csv.gz'
        file_name = f'{color}_tripdata_{year}-{month}.csv.gz'
        path = f'data/raw/{color}/{year}/{month}/{file_name}'
        print(f"Downloading: {url} -> {path}")
        !wget {url} -O {path}

        if os.path.exists(path) and os.path.getsize(path) == 0:
            print(f"Download failed: {path} is empty. Deleting the file.")
            os.remove(path)
    
```

```python
# green/yellow - 2019/2020/2021
download_file('green', '2019')
```
